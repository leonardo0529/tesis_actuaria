```python
import yfinance as yf
import pandas as pd
import numpy as np
from arch import arch_model
```

# Modelo GARCH
 

$$\epsilon_{t-p}=(r_{t}-\overline{r})$$

$$\sigma^{2}_{t-q}=\frac{\sum_{t=1}^{t}\epsilon^2_{t-p}}{n}$$



$$\sigma_{t}^{2}=w+\alpha\epsilon^{2}_{t-1}+\beta\sigma^{2}_{t-1}$$

```python
data=pd.read_excel('data_final.xlsx')
symbols=list(data['symbol'])
def rendimientos(com):return (np.diff(np.log(com))*100)
```

```python
def garch_symbol(symb):
    cierre=list(yf.download(symb,start='2019-1-1',progress=0)['Close'])
    returns=rendimientos(cierre)
    model = arch_model(returns, mean='Zero', vol='GARCH', p=1, q=1)
    model_fit = model.fit()
    w,alfa,beta=list(model_fit.params)  
    mu=np.mean(returns)
    var_hist=sum((np.array(returns)-mu)**2)/len(returns)
    pertur=(returns[-1]-mu)**2
    return  w+alfa*pertur+beta*var_hist
```

```python
varianzas=list(map(garch_symbol,symbols))
```

```python
df_varianzas=pd.DataFrame([(sym,var) for sym,var in zip(symbols,varianzas)])
df_varianzas.columns=['Symbol','Var']
df_varianzas=df_varianzas.sort_values('Var')
df_varianzas
```

```python
bajo=df_varianzas[df_varianzas['Var']<5]
medio=df_varianzas[(df_varianzas['Var']>=5) & (df_varianzas['Var']<15)]
alto=df_varianzas[df_varianzas['Var']>=15]
```
